# Mostrar imagen TIF

```python
dataset.shape
array = dataset.read(1)
pyplot.figure(figsize=(14,12))

#Mapa de colores personalizado para el categorizado
cMap = ListedColormap(['white', 'grey', 'blue','yellow','green','brown'])


#Parámetros para la leyenda
categoria = ['Urbano','Agua','Cultivo','Vegetación','Suelo desnudo/rastrojo']
colores = ['grey', 'blue','yellow','green','brown']

patches = [ pyplot.plot([],[], marker="o", ms=10, ls="", mec=None, color=colores[i], 
            label="{:s}".format(categoria[i]) )[0]  for i in range(len(categoria)) ]
pyplot.legend(handles=patches, ncol=1, numpoints=1, title='Categorías' )

pyplot.title('Ciudad de Rosario - Islas del Paraná')
#Se muestra la imagen raster clasificada
pyplot.imshow(array, cmap=cMap,aspect=1,interpolation='nearest')

```
